f(x) = 0.6x^3+5.5x^2+10x -5

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
x = Symbol("x")
f = 0.6*x**3+5.5*x**2+10*x-5
```

```python
# 1. Определяем корни 
x1, x2, x3 = solve(f,x) 
print(f"Корни находяться в точка {x1} / {x2} / {x3}")
```

```python
# 2/3. Найти интервалы, на которых функция возрастает у убывает.
#Находим производну функци
derivative_f = diff(f,x)
print(f"Производная функции: {derivative_f}")
#Находим точки экстремума
e1, e2 = solve(derivative_f, x)
y1 = f.subs(x, e1)
print(f"y1 = {y1} ")
y2 = f.subs(x, e2)
print(f"y2 = {y2} ")
print(f"т.к. в точке {e1} производная функции меняет знак с + на - точка {e1} являеться точкой максимума")
print(f"т.к. в точке {e2} производная функции меняет знак с - на + точка {e2} являеться точкой минимума")
print(f"Функции возрастает на промежутка от - бесконечности до точек с координатами от - бесокнечноти до {e1, y1} и от точки {e2, y2} до + бесконечонсти.")
print(f"Функция убывает на промежутке от точки с координатами {e1, y1} до точки {e2, y2}")
```

```python
#4. Построить график
graff = plot(f)
```

```python
#5. Вычисляем вершины
print(f"Вершинами функции будут являться точки с координатами {e1, y1} и точка с координатами {e2, y2}")
```

```python
#6/7. Определить промежутки, на котором f > 0 и f < 0
print(f"Функция f > 0 на промежутка от точки {x1} то точки {x2} и от точки {x3} до + бесконечности.")
print(f"Функция f < 0  от - бесконечности до точки {x1} и в промежутке от точки {x2} до точки {x3}.")
```
